# Tuition plotted against inflation

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

### Load Data

```python
tuition_data = pd.read_csv('tuition.csv') 
```

### Create the Dataframe 

```python
df = pd.DataFrame(tuition_data)
```

```python
df.head()
```

### Non-Resident vs Resident Tuition Increases

```python
f, ax = plt.subplots(figsize=(15, 6))

# plot non-resident tuition data
sns.set_color_codes("pastel")
sns.barplot(x="year", y="oos", data=df,
            label="Non-Resident", color="b")

# plot resident tuition data
sns.set_color_codes("muted")
sns.barplot(x="year", y="resident", data=df,
            label="Resident", color="b")

plt.xlabel('Year')
plt.ylabel('Tuition in USD')
plt.title('Tuition Increases by Year')
```

### Display out-of-state tuition increases: 1987 - 2017

```python
g = sns.barplot(x="year", y="oos", data=df, color='lightgreen')
g.set(xticklabels=[1987,1992,1997,2002,2007,2012,2017])
g.set_xticks(np.arange(0,35,5))

sns.set_style('whitegrid')

plt.xlabel('Year')
plt.ylabel('Tuition in USD')
plt.title('Out of State Tution Increases')
```

### Display resident tuition increases: 1987 - 2017

```python
g = sns.barplot(x="year", y="resident", data=df, color='lightblue')
g.set(xticklabels=[1987,1992,1997,2002,2007,2012,2017])
g.set_xticks(np.arange(0,35,5))

sns.set_style('whitegrid')

plt.xlabel('Year')
plt.ylabel('Tuition in USD')
plt.title('Resident Tution Increases')
```
